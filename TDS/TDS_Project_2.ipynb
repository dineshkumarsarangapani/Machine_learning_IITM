# Q14: Predict sentiment using Textblob. How many positive titles (original_title) are there [books.csv] ? (cut-off >0) N

## Ans: 116

```python
from textblob import TextBlob
```

```python
books_data_notna = books_data.dropna()
books_data_notna.head()
```

```python
title = pd.DataFrame()
title['original_title'] = books_data_notna['original_title']
title['original_title_Subjectivity'] = books_data_notna['original_title'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
title['original_title_Polarity'] = books_data_notna['original_title'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
title
```

```python
title[title['original_title_Polarity'] >0 ]
```
