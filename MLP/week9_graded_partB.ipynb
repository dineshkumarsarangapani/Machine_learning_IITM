```python
uri = 'https://raw.githubusercontent.com/dineshkumarsarangapani/Machine_learning_IITM/main/MLP/AirQualityUCI.csv'
```

```python
import pandas as pd

data = pd.read_csv(uri)
```

```python
data.shape
```

```python
data.head()
```

```python
data.drop(['Date', 'Time', 'Unnamed: 15', 'Unnamed: 16'], axis=1, inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data.corr()
```

```python
y = data['AH']
y
```

```python
X = data.drop(['AH'], axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,  y_train, y_test =  train_test_split(X, y, train_size=0.8, random_state=1)
```

```python
from sklearn.neural_network import MLPRegressor

regr = MLPRegressor(tol=1e-2, alpha=1e-4,solver="adam", learning_rate_init=0.1, max_iter=50, random_state=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
pipe = Pipeline([("sclae", StandardScaler()), ("mlp", regr)])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
X_test.shape
```

```python
y_pred
```

```python
pipe.score(X_test, y_test)
```

```python
pipe.score(X_train, y_train)
```
