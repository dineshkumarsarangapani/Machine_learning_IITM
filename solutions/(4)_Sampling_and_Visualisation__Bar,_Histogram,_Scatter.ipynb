# Sampling from Distributions, Bar Plots, Histograms and Scatter plots

## Import and Settings

We will import `NumPy` and `matplotlib`. In addition, we will also start with some customised layout for the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8,8]
plt.rcParams['font.size'] = 15
```

## Sampling and Plotting

`NumPy` provides a utility to generate pseudo-random numbers. We shall try to sample points from various distributions. Once we have a sample, we can then represent it pictorially using suitable plots.

### Sampling: Bernoulli

Let us generate a sample of $1000$ points from the $\text{Br}(0.7)$.

In `NumPy`:

```python
rng = np.random.default_rng(seed=100)
random_num = rng.random()
print("Random number with seed", random_num)
binomial_random = rng.choice([0,1], p = [0.4, 0.6], size = 1000)
print("Binomial Random number with size of probabilities", binomial_random)
```

### Plotting: Bar plot

Visualise the sample using a bar plot.

```python
fig, ax = plt.subplots()
plt.bar([0,1], [1000-np.count_nonzero(binomial_random), np.count_nonzero(binomial_random)])
plt.xticks([0,1]);
plt.title('Bernouli(0.6)')
```

```python
die_rand = rng.choice([1,2,3,4,5,6], p = [0.1, 0.2, 0.3, 0.2, 0.1, 0.1], size = 1000)
print("Binomial Random number with size of probabilities", die_rand)
```

```python
unique, counts = np.unique(die_rand, return_counts=True)
fig, ax = plt.subplots()
plt.bar(unique, counts)
plt.xticks(unique);
plt.title('Bernouli for a die throw!')
```

### Sampling: Gaussian

Generate a sample of $10000$ points from $\mathcal{N}(1, 4)$.

```python
x = rng.normal(1,2,size=10_000)
x
```

### Plotting: Histogram

Visualise the sample using a histogram.

```python
plt.hist(x, bins=100, edgecolor='black');
```

### Sampling: Bivariate Gaussian

Sample 1000 points from the following Bivariate Gaussian:

$$
\mathcal{N} \left( \begin{bmatrix}1 \\ 2\end{bmatrix}, \begin{bmatrix}1 & 0\\0 & 5\end{bmatrix} \right)
$$

```python
mu = np.array([1,2])
cov = np.array([[1,0], [10,5]])
x = rng.multivariate_normal(mu, cov, size = 1000)
x.shape
x[0]
```
